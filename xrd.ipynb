# Tâche 6 : xrd

Déterminez les 3 premiers pics du diffractogramme correspondant (en indiquant les indices hkl associés) pour un rayonnement X incident Cu Kα de longueur d’onde λ=1.54060 Å.

```python
import numpy as np
from mp_api.client import MPRester
from pymatgen.analysis.diffraction.xrd import *

mp_key = "XmhOs3iaP3YopOzhwHSd0LHcduZB1r1h"
mp_id = "mp-31451"
material_name = "ZrCoBi"

# Structure du matériau
with MPRester(mp_key) as m:
    prim_struc = m.get_structure_by_material_id(mp_id)
    conv_struc = SpacegroupAnalyzer(prim_struc).get_conventional_standard_structure()
    
# Rayonnement X incident CuKα
xrd = XRDCalculator(wavelength = 'CuKa1')
pattern = xrd.get_pattern(conv_struc) 
indices = pattern.hkls

# Représentation des pics du diffractograme 
xrd.get_plot(conv_struc)
```

Ce graphe représente les 3 premiers pics de diffraction pour un rayonnement X incident Cu Kα de longueur d’onde λ=1.54060 Å appliqué au ZrCoBi.


```python
print("\nPremier pic de diffraction :")
print("----------------------------\n")
print("Indices hkl associés et leur multiplicité    :", indices[0])
print("Distance interplanaire [Å]                   :", pattern.d_hkls[0])
print("2θ[°]                                        :", pattern.x[0])
print("Intensité relative                           :", pattern.y[0], "\n")

print("\nDeuxième pic de diffraction :")
print("-----------------------------\n")
print("Indices hkl associés et leur multiplicité    :", indices[1])
print("Distance interplanaire [Å]                   :", pattern.d_hkls[1])
print("2θ[°]                                        :", pattern.x[1])
print("Intensité relative                           :", pattern.y[1], "\n")

print("\nTroisième pic de diffraction :")
print("------------------------------\n")
print("Indices hkl associés et leur multiplicité    :", indices[2])
print("Distance interplanaire [Å]                   :", pattern.d_hkls[2])
print("2θ[°]                                        :", pattern.x[2])
print("Intensité relative                           :", pattern.y[2], "\n")
```

Nous avions vu suite au deuxième devoir que le type de maille du ZrCoBi est CFC (cubique face centrée). Dès lors, les hkl sont trouvées par les conditions: h+k = 2n, h+l = 2o et k+l = 2p. La multiplicité associée aux indices hkl signifie le nombre de plans différents qui donnent la même intensité.

Le 2θ quant à lui est calculé par la loi de Bragg avec θ l'angle du rayon incident.
